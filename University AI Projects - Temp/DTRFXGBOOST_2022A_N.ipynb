And then, from there, we add branch nodes, which are also known as internal nodes. All nodes must be connected, so you cannot have a tree with a node that is not connected to the rest of the tree. Each child node has one parent. So, you will notice that if you look at any single node in the diagram below, it only has one parent**

#**In contrast, parents can have more than one child, so there is no limit to a parent node's number of children. Nodes can also have no children, and in that case, they are called leaf nodes or external nodes. Trees cannot have cyclic connections, so you cannot have any loops. Everything proceeds in one direction. So, from any given node, you progress from that node, and you never end up back at that node somehow. In this case, we have the integer 25 and the root node. The integer is three, five, and seventeen in these branch nodes, and we have a value in each node.**

#**There is no reason why it must be integers. It could be other data types as well. Typically, we would only have one type of data across all the nodes, so in this case, it is all integers. Now, let us talk about different ways of characterizing the height of a tree. In this case, the height is zero, one, two, or three, so you can take the number of levels of the tree and subtract one for the root. Level one, level two, level three, level four, if you count by depth, then the root does not count, or we start indexing at zero if you want to think about it that way. # **2.Decision Trees**
#**A decision tree is a binary tree with a maximum of two, hence binary, child nodes per parent. Looking at an example tree, when you pick any given node, it splits into two or does not go anywhere. There are two main classes of decision trees. The first is a classification tree, which predicts a categorical outcome like is this a positive movie review or a negative movie review? Does the patient catch the disease or not catch the disease? The regression tree is the other main category. This is used to predict a continuous outcome like the price of a house or the temperature tomorrow. When you see the abbreviation CART in machine learning, which is abbreviated for classification and regression trees, we are talking about both kinds of decision trees.**

# **Decision tree learners attempt to find a decision rule that produces the greatest decrease in impurity at a node. While there are a number of measurements of impurity, by default DecisionTreeClassifier uses Gini impurity:**
# $G(t) = 1 - \sum\limits_{i = 1}^c {{p_i}^2}$

# **where G(t) is the Gini impurity at node t and ${p_i}$ is the proportion of observations of class c at node t. This process of finding the decision rules that create splits to decrease impurity is repeated recursively until all leaf nodes are pure (i.e., contain only one class) or some arbitrary cut-off is reached.**

# **Many decision tree model implementations use a quantity called entropy and an analysis of information gain to build up decision trees. To illustrate this process, let's consider an example. Assume that you have the following data about numbers of healthy people versus various characteristics of those people:**


#$\begin{array}{*{20}{r}}
.&{{\rm{Healthy}}}&{{\rm{Unhealthy}}}\\
{{\rm{Vegan Diet}}}&5&2\\
{{\rm{Vegetarian Diet}}}&4&1\\
{{\rm{Meat Eating Diet}}}&3&4\\
{{\rm{Age 40 +   }}}&3&5\\
{{\rm{Age  <  40}}}&9&2
\end{array}$

# **Here, we have two features in our data, Diet and Age, and we would like to build a decision tree to predict if people are healthy or not based on Diet and Age. To do this, we need to decide whether we should split our decision tree first on Age or first on Diet. Notice that we also have a total of 12 healthy people and 7 unhealthy people represented in the data.To begin, we will calculate the overall or total entropy of the classes in our data. This is defined as follows:**

# ${\rm{E}} =  - {p_1}{\log _2}({p_1}) - {p_2}{\log _2}({p_2}) - {p_3}{\log _2}({p_3})$

# ${p_1}$, ${p_2}$ , ${p_3}$ **and so on, are the probabilities of a first category, a second category, and so on. In our particular case (because we have 12 healthy people and 7 unhealthy people), our total entropy is as follows:**


#${\rm{E(Health)}} =  - \left( {\frac{{12}}{{19}}} \right){\log _2}\left( {\frac{{12}}{{19}}} \right) - \left( {\frac{7}{{19}}} \right){\log _2}\left( {\frac{7}{{19}}} \right) = 0.95$

#**This 0.95 measure represents the homogeneity of our health data. It goes between 0 and 1, with high values corresponding to less homogeneous data.**

# **Now to determine whether we should split our tree first on Age or first on Diet, we will calculate which of these features gives us the most information gain. Simply put, we will find the feature that gives us the most homogeneity after splitting on that feature, as measured by the preceding entropy. This decrease in entropy is called information gain.**

#**The information gain for a certain feature in our example is defined as follows:**

# ${\rm{G = E(Health) - E(Health,Feature)}}$

#**Here, E(Health, Feature) is a second measure of entropy with respect to the given feature (Age or Diet). For Diet, this second measure would be calculated as follows:**


# ${\rm{E(Health, Age) = }}{{\rm{p}}_{{\rm{40}}}}{\rm{ + E(Health,40 + ) +  }}{{\rm{p}}_{{\rm{ < 40}}}}{\rm{ E( Health,  < 40)}}$

# **Note that the quantities p40+ and p<40 are the probabilities of having an age of 40+ or <40 (8/19 and 11/19, respectively). The quantities E(Health,40+) and E(Health,<40) are the health entropies (as defined in the preceding formula) but only using the counts corresponding to those Age 40+ and Age <40, respectively.**

#**For our example data, the information gain for the Age feature comes out to 0.152 and the information gain for the Diet feature comes out to 0.079. Thus, we would choose to split our decision tree on the Age feature first because it increases the overall homogeneity of our data the most.**

# **Random Forest**
#**Decision trees are prone to over-fitting to the training data and not generalizing well to data that the model hasn't seen before. The solution to this problem is random forests. So random forests are ensembles of decision trees of multiple decision trees, and they tend to outperform any single decision tree model. They are also less prone to over-fitting relative to a single decision tree. Ultimately, random forests consist of just a whole bunch of decision trees in them. We could repeat a stochastic process. So, a process that includes randomness, random forest a hundred times. We use sampling with replacement to sample some data from all of our training data. And this gives us a new sample of data. If we grab something out of our sample set, we put it back in after we've sampled it. So, you could end up with a sample of the same data point multiple times in a sample. And we call this a bootstrap aggregate or bagged sample when we do it that way.**
#**Once we have our bagged sample, we then fit a decision tree. And when we split at each node, we might use all the features available to us, or we might randomly subset from the complete feature set. Random forests are stochastic in two ways. There are two stochastic with respect to how we bag samples from our entire training dataset. And they are also random in the sense that we can be using subsets of our complete feature set. We would then repeat that process many times, say a hundred times to get an entire random forest of decision trees. After we've done that, say a hundred times, we average our diverse decision trees together, which tends to reduce variance. So we tend to get more reliable model outputs instead of more widely varying model outputs. So our model then tends to perform better on data that it hasn't seen before. It avoids over-fitting.**

# **Gradient-Boosted Trees**

# **Gradient-Boosted Tree (GBT) is another tree-based ensemble algorithm similar to Random Forest. GBTs use a technique known as boosting to create a strong learner from weak learners (shallow trees). GBTs train an ensemble of decision trees sequentially, with each succeeding tree decreasing the error of the previous tree. This is done by using the residuals of the previous model to fit the next model. This residual-correction process is performed a set number of iterations with the number of iterations determined by cross-validation until the residuals have been fully minimized.**

# **1.Import the necessary libraries for Gradient-Boosted Tree** 

```python
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.metrics import mean_absolute_error
```

```python
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target)
```

# **2.We do train test split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# **3. We use two nodes with three decision trees in GradientBoostingRegressor**

```python
regressor = GradientBoostingRegressor(
    max_depth=2,
    n_estimators=3,
    learning_rate=1.0
)
regressor.fit(X_train, y_train)
```

# **4. We compute the errors and the best estimators.**

```python
errors = [mean_squared_error(y_test, y_pred) for y_pred in regressor.staged_predict(X_test)]
best_n_estimators = np.argmin(errors)
```

# **5. We use the best estimators and create our model.**

```python
best_regressor = GradientBoostingRegressor(
    max_depth=2,
    n_estimators=best_n_estimators,
    learning_rate=1.0
)
best_regressor.fit(X_train, y_train)
```

# **6. We predict our test set**


```python
y_pred = best_regressor.predict(X_test)
```

# **7. compute the mean absolute error.**

```python
mean_absolute_error(y_test, y_pred)
```

# **XGBoost**

# **XGBoost is one of the best gradient-boosted tree implementations currently available. Released on March 27, 2014, by Tianqi Chen as a research project, XGBoost has become the dominant machine learning algorithm for classification and regression. XGBoost was designed using the general principles of gradient boosting, combining weak learners into a strong learner. But while gradient-boosted trees are built sequentially – slowly learning from data to improve its prediction in succeeding iteration, XGBoost builds trees in parallel.XGBoost produces better prediction performance by controlling model complexity and reducing overfitting through its built-in regularization. XGBoost uses an approximate algorithm to find split points when finding the best split points for a continuous feature. The approximate splitting method uses discrete bins to bucket continuous features, significantly speeding up model training. XGBoost includes another tree growing method using a histogram-based algorithm which provides an even more efficient method of bucketing continuous features into discrete bins. But while the approximate method creates a new set of bins per iteration, the histogram-based approach reuses bins over multiple iterations. This approach allows for additional optimizations that are not achievable with the approximate method, such as the ability to cache bins and parent and sibling histogram subtraction. To optimize sorting operations, XGBoost stores sorted data in in-memory units of blocks. Sorting blocks can be efficiently distributed and performed by parallel CPU cores. XGBoost can effectively handle weighted data via its weighted quantile sketch algorithm, can efficiently handle sparse data, is cache-aware, and supports out-of-core computing by utilizing disk space for large datasets, so data does not have to fit in memory.**
