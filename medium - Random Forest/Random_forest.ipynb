# RANDOM FOREST
in this notebook we implement an example of how to use random forest

## Imports

***general imports***

```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
import matplotlib.pyplot as plt 
import seaborn as sns 
import random 
```

***sklearn imports***

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

## Load Data

***we use iris dataset from sklearn***

```python
iris = sns.load_dataset("iris")

iris
```

***draw a scatter plot***

```python
_ = sns.scatterplot(x='sepal_width', y='petal_length', hue='species', data=iris)
```

***creating the feature array***

```python
X = iris[["sepal_width", "petal_length"]]
```

***creating the target array***

```python
y = iris.species
```

## Pre-Process Data

***splitting the data***

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Load and Fit Model

***fitting a decision tree model***

```python
iris_dt = DecisionTreeClassifier().fit(X_train, y_train)
```

***fitting a random forest model***

```python
rf_model = RandomForestClassifier(n_estimators=100)

rf_model.fit(X_train, y_train)
```

## Model Evaluation

***evaluating decision tree performance***

```python
iris_dt_yhat = iris_dt.predict(X_test)

accuracy_score(iris_dt_yhat, y_test) 
```

***evaluating random forest performance***

```python
rf_yhat = rf_model.predict(X_test)

accuracy_score(rf_yhat, y_test)
```
